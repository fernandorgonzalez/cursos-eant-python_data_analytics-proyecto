```python
#Importamos librerías
import emoji
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import pandas as pd
import seaborn as sns
```

```python
#Archivo donde tomamos el set de datos para bicicletas luego de su limpieza y normalización con la notebook bicicletas.ipynb
input_file_name_bicicletas = "https://github.com/fernandorgonzalez/cursos-eant-python_data_analytics-proyecto/blob/main/bicicletas.xlsx?raw=true"
input_sheet_name_bicicletas = "bicicletas"
```

```python
#Archivo donde tomamos el set de datos para contaminantes luego de su limpieza y normalización con la notebook contaminantes.ipynb
input_file_name_contaminantes = "https://github.com/fernandorgonzalez/cursos-eant-python_data_analytics-proyecto/blob/main/contaminantes.xlsx?raw=true"
input_sheet_name_contaminantes = "contaminantes"
```

```python
#Archivo donde tomamos el set de datos para covid luego de su limpieza y normalización con la notebook covid.ipynb
input_file_name_covid = "https://github.com/fernandorgonzalez/cursos-eant-python_data_analytics-proyecto/blob/main/covid.xlsx?raw=true"
input_sheet_name_covid = "covid"
```

```python
#Archivo donde tomamos el set de datos para subte luego de su limpieza y normalización con la notebook subte.ipynb
input_file_name_subte = "https://github.com/fernandorgonzalez/cursos-eant-python_data_analytics-proyecto/blob/main/subte.xlsx?raw=true"
input_sheet_name_subte = "subte"
```

```python
#Archivo donde tomamos el set de datos para vehículos luego de su limpieza y normalización con la notebook vehiculos.ipynb
input_file_name_vehiculos = "https://github.com/fernandorgonzalez/cursos-eant-python_data_analytics-proyecto/blob/main/vehiculos.xlsx?raw=true"
input_sheet_name_vehiculos = "vehiculos"
```

```python
#Carga de dataframe para bicicletas
df_bicicletas = pd.read_excel(input_file_name_bicicletas, sheet_name = input_sheet_name_bicicletas)
df_bicicletas
```

```python
#Carga de dataframe para contaminantes
df_contaminantes = pd.read_excel(input_file_name_contaminantes, sheet_name = input_sheet_name_contaminantes)
df_contaminantes
```

```python
#Carga de dataframe para Casos de Covid
df_covid = pd.read_excel(input_file_name_covid, sheet_name = input_sheet_name_covid)
df_covid
```

```python
#Carga de dataframe para Subte
df_subte = pd.read_excel(input_file_name_subte, sheet_name = input_sheet_name_subte)
df_subte
```

```python
#Carga de dataframe para Vehículos
df_vehiculos = pd.read_excel(input_file_name_vehiculos, sheet_name = input_sheet_name_vehiculos)
df_vehiculos
```

```python
#Concatenamos los cuatro datasets
df = pd.concat([df_bicicletas,df_contaminantes,df_covid,df_subte,df_vehiculos]).reset_index(drop=True)
df
```

```python
#Previsualización de la paleta a utilizar para los meses del año
sns.color_palette('plasma',12)
```

```python
#Definición de la paleta para los meses del año
paleta_meses = {}
for i in range(0,12):
    paleta_meses[i+1] = sns.color_palette('plasma',12)[i]
paleta_meses
```

```python
#Previsualización de la paleta a utilizar para la comparación de los indicadores
sns.color_palette()
```

```python
#Definición de la paleta para la comparación de indicadores
paleta_comparacion = {}
for i in range(0,len(df['Tipo'].value_counts().index)):
    paleta_comparacion[df['Tipo'].value_counts().index[i]] = sns.color_palette()[i]
paleta_comparacion
```

```python
#Visualización de los Casos de Covid
plt.figure(figsize=(10,7))
s = sns.barplot(data = df_covid,
            x = 'Mes',
            y = 'Total',
            palette = paleta_meses)
for p in s.patches:
    s.annotate(int(p.get_height()), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=10,
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   weight="bold",color="#565656")
plt.title("Casos de Covid 2020", fontsize = 20,backgroundcolor="#565656",
  weight="bold",color="white",style="italic",loc="center",pad=30)
plt.xlabel('Mes',weight="bold",size=20,color="#565656")
plt.ylabel('Total',weight="bold",size=20,color="#565656")
plt.tick_params(labelsize=10,length=0)
plt.box(False)
plt.grid(linewidth=0.5,color="grey",linestyle="-.")
plt.show()
```

```python
#Visualización de los Contaminantes
plt.figure(figsize=(10,7))
s = sns.barplot(data = df_contaminantes,
            x = 'Mes',
            y = 'Total',
            palette = paleta_meses)
for p in s.patches:
    s.annotate(p.get_height(), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=10,
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   weight="bold",color="#565656")
plt.title("Contaminantes 2020 (" + df_contaminantes.iloc[0][5] + ")", fontsize = 20,backgroundcolor="#565656",
  weight="bold",color="white",style="italic",loc="center",pad=30)
plt.xlabel('Mes',weight="bold",size=20,color="#565656")
plt.ylabel('Total',weight="bold",size=20,color="#565656")
plt.tick_params(labelsize=10,length=0)
plt.box(False)
plt.grid(linewidth=0.5,color="grey",linestyle="-.")
plt.show()
```
